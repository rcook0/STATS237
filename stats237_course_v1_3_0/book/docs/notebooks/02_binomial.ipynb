# Binomial (CRR)

Demonstrates European vs American pricing and a simple replication intuition.

```python
from stats237_quantlib.pricing.binomial import crr_european, crr_american

S=100; K=100; r=0.02; q=0.0; sigma=0.2; T=1.0; N=100

call_e = crr_european(S,K,r,q,sigma,T,N, option_type='call')
put_a  = crr_american(S,K,r,q,sigma,T,N, option_type='put')

(call_e, put_a)
```
